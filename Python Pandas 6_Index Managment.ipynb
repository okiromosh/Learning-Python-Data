```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.sample(3)
```

Creating a name for the index col

```python
df.index.name = 'job_index'
```

**reset_index()**

creates new index for a filtered object

```python
df_usa = df[df['job_country'] == 'United States']
```

```python
df_usa.head(10)
```

```python
df_usa.reset_index(inplace=True)
df_usa
```

**set_index()**

Sets a selected column as the index

```python
'''
In this case we want the original job_index back 
instead of the new index we created
'''
df_usa.set_index('job_index', inplace=True)
```

```python
df_usa
```

**sort_index()**

useful to sort index that dont have labels

```python
median_pivot = df_usa.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median','max','min'])
median_pivot
```

```python
'''
this sorts the index 'job_title_short' in alphabetical
'''
median_pivot.sort_index()
```
