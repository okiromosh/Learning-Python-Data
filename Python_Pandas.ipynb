# **Cleaning Data**

```python
import pandas as pd
from datasets import load_dataset

# Loading Data

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
# Forcusing on date first
df.info() # <job_posted_date> seen as object not datetime
```

```python
df.job_posted_date[0]
```

```python
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
df.info()
```

```python
df.job_posted_date.dt.month
```

**Creating a new column to display months only**

```python
df['job_created_month'] = df.job_posted_date.dt.month
```

```python
df.info()
```

```python
df
```

**Sorting post date**

```python
df.sort_values('job_posted_date', inplace=True) # inplace=True makes the sorting changes directly to the dataframe
```

```python
df
```

**Dropping/Deleting a column**

```python
df.drop(labels='salary_hour_avg', axis=1, inplace=True) #axis=1 deletes the column label too
```

**Removing null/none/na values**

```python
df.dropna(subset=['salary_year_avg'], inplace=True)
```

```python
df.info()
```

# **Analyzing Data**

```python
import pandas as pd
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Cleaning data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['salary_year_avg'].min()
```

```python
min_salary = df['salary_year_avg'].idxmin() # Finding the index of the least salary
```

```python
df.iloc[min_salary]
```
